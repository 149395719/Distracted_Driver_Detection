读取driver_imgs_list文件

```python
def get_driver_data():
    dr_dict = {}
    dr_set = set()
    path = os.path.join('driver_imgs_list.csv')
    print('Read drivers data')
    f = open(path, 'r')
    line = f.readline()
    while (1):
        line = f.readline()
        if line == '':
            break
        arr = line.strip().split(',')   
        dr_dict[arr[2]] = arr[0]
        dr_set.add(arr[0])
    f.close()
    print(len(dr_dict),dr_set)
    return dr_dict,dr_set
```
